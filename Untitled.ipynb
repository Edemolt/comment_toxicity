```python
import tensorflow as tf
from tensorflow.keras.layers import TextVectorization

def prepare_dataset(df, text_column, label_column, max_features=1000000, output_sequence_length=1800, batch_size=16, buffer_size=16000):
    """
    Prepares a TensorFlow dataset from a DataFrame.

    Parameters:
    - df (pd.DataFrame): The input DataFrame containing text and labels.
    - text_column (str): The name of the column containing the text data.
    - label_column (str): The name of the column containing the labels.
    - max_features (int): The maximum number of unique words to keep.
    - output_sequence_length (int): The length of the output sequences after padding/truncating.
    - batch_size (int): The batch size for the dataset.
    - buffer_size (int): The buffer size for shuffling the dataset.

    Returns:
    - tf.data.Dataset: The processed TensorFlow dataset.
    """
    
    # Extract text and labels from the DataFrame
    x = df['comment_text']
    y = df[df.columns[2:]]

    # Create a TextVectorization layer
    vectorizer = TextVectorization(max_tokens=MAX_FEATURES, output_sequence_length=output_sequence_length, output_mode='int')

    # Adapt the vectorizer to the text data
    vectorizer.adapt(x.values)

    # Vectorize the text data
    vectorized_text = vectorizer(x.values)

    # Create the TensorFlow dataset
    dataset = tf.data.Dataset.from_tensor_slices((vectorized_text, y))
    dataset = dataset.cache()
    dataset = dataset.shuffle(buffer_size)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)  # Automatically chooses a good prefetch buffer size

    return dataset
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the data
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_test_labels = pd.read_csv('test_labels.csv')
# Prepare training dataset

train = prepare_dataset(df_train, text_column='text', label_column='label')

df_test = pd.merge(df_test, df_test_labels, on='id')
# Split test data into validation and test sets
test_df, val_df = train_test_split(df_test, test_size=0.3, random_state=42, shuffle=True)

# Prepare test and validation datasets
val = prepare_dataset(val_df, text_column='text', label_column='label')
test = prepare_dataset(test_df, text_column='text', label_column='label')

# Now you have `train_dataset`, `val_dataset`, and `test_dataset` ready for use
```

# Making and training the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding
```

```python
model = Sequential()

model.add(Embedding(MAX_FEATURES + 5, 32))
# 1 embedding per word, will ahve 32 values log
model.add(Bidirectional(LSTM(32, activation = 'tanh')))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(256, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(6, activation = 'sigmoid'))
```

```python
model.compile( loss= 'BinaryCrossentropy', optimizer= 'Adam', metrics = ["accuracy"])
```

```python
model.summary()
```

```python
#  train time :)
history = model.fit( train, epochs = 5, validation_data = val)
```

```python
# Save the model to a file
model.save('new_model.h5')

print("Model saved successfully!")
```
